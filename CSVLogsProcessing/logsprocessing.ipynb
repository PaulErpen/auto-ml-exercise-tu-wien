```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

# Elastic Net

```python
beijing_elasticnet = pd.read_csv(
    "../output/beijing/ElasticNetManager.csv", 
    names=["timestamp","temperature","rmse","alpha","l1","is_best","is_current"])
display(beijing_elasticnet)
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(beijing_elasticnet["alpha"], beijing_elasticnet["rmse"], c=beijing_elasticnet["l1"])
plt.title("Elastic Net - Parameter tuning")
ax.legend(*scatter.legend_elements(), title="L1 Ratio", loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel("Alpha")
plt.ylabel("MSE")
plt.show()
```

```python
beijing_current = beijing_elasticnet[beijing_elasticnet["is_current"] == True]
beijing_best = beijing_elasticnet[beijing_elasticnet["is_best"] == True]
plt.scatter(beijing_elasticnet["timestamp"], beijing_elasticnet["rmse"], c="blue")
plt.scatter(beijing_current["timestamp"], beijing_current["rmse"], c="orange")
plt.scatter(beijing_best["timestamp"], beijing_best["rmse"], c="red")
plt.title("Elastic Net - Parameter tuning")
plt.xlabel("Time since start in seconds")
plt.ylabel("MSE")
plt.legend(title="Result type", handles=[patches.Patch(color="blue", label="Normal"), patches.Patch(color="red", label="Best"), patches.Patch(color="orange", label="current")])
```

# Neural Network

```python
beijing_mlp = pd.read_csv(
    "../output/beijing/MlpManager.csv", 
    names=["timestamp","temperature","rmse","n_neurons_l1","n_neurons_l2","alpha","l1","is_best","is_current"])
display(beijing_mlp)
```

```python
plt.scatter(beijing_mlp["timestamp"], beijing_mlp["n_neurons_l1"], c="orange")
plt.scatter(beijing_mlp["timestamp"], beijing_mlp["n_neurons_l2"], c="blue")
plt.legend(title="Number of neurons", handles=[patches.Patch(color="orange", label="Layer 1"), patches.Patch(color="blue", label="Layer 2")])
plt.title("MLP - Number of Neurons on hidden layers")
plt.xlabel("time since start in seconds")
plt.ylabel("number of neurons")
```
